```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#Time series forecasting we need index column as time 
df = pd.read_csv('AirPassengers.csv' , parse_dates=['Month'])
df.index = df['Month']
df.drop( columns = ['Month'], inplace = True)
```

```python
# Import modules for testing 
from statsmodels.tsa.stattools import adfuller 
rolmean = df.Passengers.rolling(window=12).mean()
rolstd = df.Passengers.rolling(window=12).std()
rolmean.head(20),rolstd.head(20)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(window=12).mean()
    rolstd = timeseries.rolling(window=12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
#plotting the mean and std with the df 
#Plotting the original data
test_stationarity(df)
```

```python
#Estimating and eliminating the Trend
df_log = np.log(df)
f , axs = plt.subplots(1 , 2 ,figsize=(15,5) )
axs[0].set_title('Original')
axs[0].plot(df , color ='blue')
axs[1].set_title('Log Of Original')
axs[1].plot(df_log ,color= 'red')
```

```python
#plotting the moving average of df_log
df_log_moving_avg = df_log.rolling(window = 12).mean()
plt.plot(df_log)
plt.plot(df_log_moving_avg, color='red')
```

```python
df_log_moving_avg_diff = df_log - df_log_moving_avg
df_log_moving_avg_diff.head(12)
```

```python
#plotting the df_log 
df_log_moving_avg_diff.dropna(inplace=True)
test_stationarity(df_log_moving_avg_diff)
```

```python
#Exponentially weighted moving average
expwighted_avg = df_log.ewm(halflife=12).mean()
plt.plot(df_log)
plt.plot(expwighted_avg, color='red')
```

```python
df_log_ewma_diff = df_log - expwighted_avg
test_stationarity(df_log_ewma_diff)
```

```python
df_log_diff = df_log - df_log.shift()
plt.plot(df_log_diff)
```

```python
df_log_diff.dropna(inplace=True)
test_stationarity(df_log_diff)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df_log)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(df_log, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

```python
ts_log_decompose = residual
ts_log_decompose.dropna(inplace=True)
test_stationarity(ts_log_decompose)
```

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf
```

```python
lag_acf = acf(df_log_diff, nlags=20)
lag_pacf = pacf(df_log_diff, nlags=20, method='ols')
```

```python
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_log_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')

#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_log_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model = ARIMA(df_log, order=(2, 1, 0))  
results_AR = model.fit(disp=-1)  
plt.plot(df_log_diff)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-df_log_diff)**2))
```

```python
model = ARIMA(df_log, order=(0, 1, 2))  
results_MA = model.fit(disp=-1)  
plt.plot(df_log_diff)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-df_log_diff)**2))
```

```python
model = ARIMA(df_log, order=(2, 1, 2))  
results_ARIMA = model.fit(disp=-1)  
plt.plot(df_log_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-df_log_diff)**2))
```

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
predictions_ARIMA_diff.head()
```

```python
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
predictions_ARIMA_diff_cumsum.head()
```

```python
predictions_ARIMA_log = pd.Series(df_log.iloc[0], index=df_log.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA_log.head()
```

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(df)
plt.plot(predictions_ARIMA)
plt.title('RMSE: %.4f'%(np.sqrt(sum((predictions_ARIMA-df)**2)/len(df))))
```
